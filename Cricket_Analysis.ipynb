# IPL Analysis


# Highest Runs

```python
player["batsman_runs"].sum().sort_values(ascending=False).head(10).reset_index()
```

# most no of sixes

```python
sixes=ipl[ipl["batsman_runs"]==6]
sixesg=sixes.groupby("batsman")
```

```python
sixesg["batsman_runs"].count().sort_values(ascending=False).reset_index()
```

```python


temp1=ipl[ipl["over"]>15]
temp1=temp1[(temp1["batsman_runs"]==6)|(temp1["batsman_runs"]==4)]
```

# most no 6 and 4 in last 5 overs(best finisher)

```python
temp2=temp1.groupby("batsman")
temp2["batsman_runs"].count().sort_values(ascending=False).head(10).reset_index()
```

## Kholis highest runs against any match in ipl

```python
temp3=ipl[ipl["batsman"]=="V Kohli"]
```

```python
temp3.groupby("match_id")["batsman_runs"].sum().sort_values(ascending=False).reset_index()
```
